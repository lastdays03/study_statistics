## 1. Sigma ($\sum$) = `sum()` or `for loop`

$$ y = \sum_{i=1}^{n} x_i $$

ì´ ìˆ˜ì‹ì€ "ë¦¬ìŠ¤íŠ¸ $x$ì— ìˆëŠ” ê±° $1$ë²ˆë¶€í„° $n$ë²ˆê¹Œì§€ ì‹¹ ë‹¤ ë”í•´ë¼"ë¼ëŠ” ëœ»ì…ë‹ˆë‹¤.

```python
x = [1, 2, 3, 4, 5]

# ìˆ˜í•™: Sigma x_i
# íŒŒì´ì¬: sum(x)
y_sigma = sum(x)

# íŒŒì´ì¬: for loop
y_loop = 0
for val in x:
    y_loop += val
    
print(f"Sigma Result: {y_sigma}")
print(f"Loop Result: {y_loop}")
```

## 2. Pi ($\prod$) = `math.prod()`

$$ y = \prod_{i=1}^{n} x_i $$

ë”í•˜ê¸°($\sum$) ë§ê³  ê³±í•˜ê¸°($\times$)ë¥¼ í•˜ë¼ëŠ” ëœ»ì…ë‹ˆë‹¤. í™•ë¥  ê³„ì‚°í•  ë•Œ ë§ì´ ì”ë‹ˆë‹¤.

```python
import math

x = [1, 2, 3, 4, 5]

# ìˆ˜í•™: Pi x_i
# íŒŒì´ì¬: math.prod(x)
y_pi = math.prod(x)

# íŒŒì´ì¬: for loop
y_loop = 1
for val in x:
    y_loop *= val
    
print(f"Pi Result: {y_pi}")
print(f"Loop Result: {y_loop}")
```

## 3. Logarithm ($\log$): ê³±ì…ˆì„ ë§ì…ˆìœ¼ë¡œ

í° ìˆ˜ë¥¼ ë‹¤ë£° ë•Œ, í˜¹ì€ ê³±í•˜ê¸° ê³„ì‚°ì´ ë„ˆë¬´ ë³µì¡í•  ë•Œ ë¡œê·¸ë¥¼ ì”Œì›ë‹ˆë‹¤.
$$ \log(a \times b) = \log(a) + \log(b) $$

> **Why?**: í™•ë¥  $0.0001 \times 0.0001$ ê³„ì† í•˜ë©´ ì–¸ë”í”Œë¡œìš°(0ì´ ë¨)ê°€ ë‚©ë‹ˆë‹¤. ë¡œê·¸ë¥¼ ì”Œìš°ë©´ ë§ì…ˆì´ ë˜ì–´ ì•ˆì „í•©ë‹ˆë‹¤.

```python
a = 1e-10
b = 1e-10

print(f"Just Multiply: {a * b}") # ë„ˆë¬´ ì‘ì•„ì§
print(f"Log Sum: {math.log(a) + math.log(b)}") # ì•ˆì „í•œ ìˆ«ì ë²”ìœ„
```

# Part 2. ğŸ’» Manual Implementation: ë¯¸ë¶„ (Derivative)

ë¯¸ë¶„ ê³µì‹ì„ ì™¸ìš¸ í•„ìš” ì—†ìŠµë‹ˆë‹¤. **"ì•„ì£¼ ì¡°ê¸ˆ ì›€ì§ì˜€ì„ ë•Œ, ì–¼ë§ˆë‚˜ ë³€í–ˆë‚˜?"** ë§Œ ì•Œë©´ ë©ë‹ˆë‹¤.

$$ f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h} $$

```python
def numerical_derivative(f, x):
    h = 1e-4 # 0.0001 (ì•„ì£¼ ì‘ì€ ê°’)
    return (f(x + h) - f(x)) / h

# ì‹¤í—˜ ëŒ€ìƒ í•¨ìˆ˜: y = x^2
def my_func(x):
    return x**2

# x=3ì¼ ë•Œ ë¯¸ë¶„ê°’ (ê³µì‹ìœ¼ë¡œëŠ” 2xë‹ˆê¹Œ 6ì´ ë‚˜ì™€ì•¼ í•¨)
slope = numerical_derivative(my_func, 3.0)
print(f"f(x) = x^2, x=3ì¼ ë•Œ ê¸°ìš¸ê¸°: {slope:.4f}")
```

### Gradient Descent (ê²½ì‚¬í•˜ê°•ë²•) ì§ì ‘ êµ¬í˜„í•˜ê¸°
ê¸°ìš¸ê¸°ë¥¼ ì•Œë©´, ì‚°ì„ ë‚´ë ¤ì˜¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤. (í•¨ìˆ˜ì˜ ìµœì†Ÿê°’ ì°¾ê¸°)

```python
# ì‹œì‘ì  (ëœë¤)
x = 10.0
learning_rate = 0.1

print(f"Start x: {x}")

for i in range(20):
    slope = numerical_derivative(my_func, x)
    x = x - learning_rate * slope # ê²½ì‚¬ ë°˜ëŒ€ ë°©í–¥ìœ¼ë¡œ ì´ë™
    print(f"Step {i+1}: x = {x:.4f}, slope = {slope:.4f}")

print("Result: 0ì— ê°€ê¹Œì›Œì¡Œë‚˜ìš”?")
```

# Part 4. ğŸ§ª Experiment: ì»´í“¨í„°ì˜ í•œê³„

### Scenario: ë¯¸ë¶„í•  ë•Œ `h`ê°€ ë„ˆë¬´ ì‘ìœ¼ë©´?
ìˆ˜í•™ì ìœ¼ë¡œëŠ” $h \to 0$ì¼ìˆ˜ë¡ ì •í™•í•˜ì§€ë§Œ, ì»´í“¨í„°ëŠ” ì†Œìˆ˜ì  ì •ë°€ë„(Float Precision) í•œê³„ê°€ ìˆìŠµë‹ˆë‹¤.

```python
true_slope = 6.0 # x=3ì¼ ë•Œ ì‹¤ì œ ë¯¸ë¶„ê°’ (2x = 6)
h_values = [1e-1, 1e-4, 1e-7, 1e-15, 1e-16]

print(f"True Slope: {true_slope}")
print("-"*30)

for h in h_values:
    estimated_slope = (my_func(3.0 + h) - my_func(3.0)) / h
    error = abs(true_slope - estimated_slope)
    print(f"h = {h:.1e} -> Slope: {estimated_slope}, Error: {error:.5f}")

print("-"*30)
print("hê°€ ë„ˆë¬´ ì‘ìœ¼ë©´(1e-16) ì»´í“¨í„°ê°€ ë§ì…ˆì„ ë¬´ì‹œí•´ì„œ 0ì´ ë˜ì–´ë²„ë¦½ë‹ˆë‹¤(Numerical Error). 1e-4 ì •ë„ê°€ ì ë‹¹í•©ë‹ˆë‹¤.")
```
