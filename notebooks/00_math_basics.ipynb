## 1. Sigma ($\sum$) = `for` Loop

$$ y = \sum_{i=1}^{n} (x_i + 10) $$

ì´ ê¸°í˜¸ëŠ” ë¬´ì„­ê²Œ ìƒê²¼ì§€ë§Œ, ì‚¬ì‹¤ **"ë¦¬ìŠ¤íŠ¸ ëŒë©´ì„œ ë‹¤ ë”í•´ë¼"**ë¼ëŠ” ëœ»ì…ë‹ˆë‹¤.

### ğŸ“ Math Example
ë°ì´í„°ê°€ $[1, 2, 3]$ ì´ë¼ê³  í•©ì‹œë‹¤ ($n=3$).
$$ \sum_{i=1}^{3} (x_i + 10) = (1+10) + (2+10) + (3+10) $$
$$ = 11 + 12 + 13 = 36 $$

```python
data = [1, 2, 3] # ì˜ˆì œ ë°ì´í„°

# 1. ì´ŒìŠ¤ëŸ¬ìš´ ë°©ì‹ (C-style)
total = 0
for x in data:
    total += (x + 10)
print(f"For Loop: {total}")

# 2. íŒŒì´ì¬ìŠ¤ëŸ¬ìš´ ë°©ì‹ (Pythonic)
total_pythonic = sum([x + 10 for x in data])

print(f"Pythonic Sum: {total_pythonic}")
```

## 3. Logarithm ($\log$): ê³±ì…ˆì„ ë§ì…ˆìœ¼ë¡œ

ë¡œê·¸ëŠ” ê³±ì…ˆì„ ë§ì…ˆìœ¼ë¡œ ë°”ê¿”ì£¼ëŠ” ë§ˆë²•ì˜ ë„êµ¬ì…ë‹ˆë‹¤.

$$ \log(A \times B) = \log(A) + \log(B) $$

### ğŸ“ Math Example
ìƒìš©ë¡œê·¸(ë°‘ì´ 10)ë¼ê³  ê°€ì •í•´ë´…ì‹œë‹¤.
$$ A = 100 (=10^2), \quad B = 1000 (=10^3) $$
$$ \log(100 \times 1000) = \log(100000) = 5 $$
$$ \log(100) + \log(1000) = 2 + 3 = 5 $$
ê²°ê³¼ê°€ ê°™ìŠµë‹ˆë‹¤! ê³±í•˜ê¸°(í° ìˆ˜)ë¥¼ ë”í•˜ê¸°(ì‘ì€ ìˆ˜)ë¡œ ë°”ê¿€ ìˆ˜ ìˆì£ .

```python
import math

# ìì—°ë¡œê·¸(ln) ê¸°ì¤€
prob_a = 1e-10
prob_b = 1e-10

print(f"ê·¸ëƒ¥ ê³±í•˜ê¸°: {prob_a * prob_b}") # ë„ˆë¬´ ì‘ì•„ì„œ 0 ì·¨ê¸‰ë‹¹í•  ìˆ˜ ìˆìŒ
print(f"ë¡œê·¸ í•©: {math.log(prob_a) + math.log(prob_b)}") # -46.05... (ì•ˆì „í•œ ìˆ«ì ë²”ìœ„)
# math.logì˜ ê¸°ë³¸ ë°‘(base)ì€ e (ìì—°ìƒìˆ˜)ì…ë‹ˆë‹¤.
# ìì—°ìƒìˆ˜ eëŠ” ë¯¸ë¶„ ì‹œ ê³„ì‚°ì´ ê°€ì¥ ê°„ê²°í•˜ë©°, ì •ê·œë¶„í¬ ë“± ë§ì€ í†µê³„ì  ê³µì‹ì˜ ê¸°ì´ˆê°€ ë˜ê¸° ë•Œë¬¸ì— ê¸°ë³¸ê°’ìœ¼ë¡œ ì‚¬ìš©ë©ë‹ˆë‹¤.
print(f"math.logì˜ ê¸°ë³¸ ë°‘(base): {math.e}")
print(f"math.log(math.e): {math.log(math.e)}")
```

# Part 2. ë¯¸ë¶„ (Calculus): ë³€í™”ë¥¼ ë‹¤ë£¨ëŠ” ê¸°ìˆ 

ë¯¸ë¶„($f'$)ì€ **"ì…ë ¥ì„ ì•„ì£¼ ì¡°ê¸ˆ($h$) ë°”ê¿¨ì„ ë•Œ ê²°ê³¼ê°€ ì–¼ë§ˆë‚˜ ë³€í•˜ë‚˜?"** ì…ë‹ˆë‹¤.

$$ f'(x) \approx \frac{f(x+h) - f(x)}{h} $$

### ğŸ“ Math Example
í•¨ìˆ˜ê°€ $f(x) = x^2$ ì´ê³ , $x=3$ ì—ì„œì˜ ê¸°ìš¸ê¸°ë¥¼ êµ¬í•˜ê³  ì‹¶ìŠµë‹ˆë‹¤.
ì•„ì£¼ ì‘ì€ ìˆ˜ $h=0.001$ ì„ ì¨ë´…ì‹œë‹¤.

1. $f(3) = 3^2 = 9$
2. $f(3+h) = f(3.001) = 3.001^2 = 9.006001$
3. ë³€í™”ëŸ‰ = $9.006001 - 9 = 0.006001$
4. ë³€í™”ìœ¨(ê¸°ìš¸ê¸°) = $\frac{0.006001}{0.001} \approx 6.0$

ì •ë‹µ $2x = 6$ ê³¼ ê±°ì˜ ê°™ìŠµë‹ˆë‹¤!

```python
def numerical_derivative(f, x):
    h = 1e-4 # 0.0001 (ì•„ì£¼ ì‘ì€ ë³€í™”)
    return (f(x + h) - f(x)) / h

# ì‹¤í—˜ ëŒ€ìƒ: y = x^2
slope = numerical_derivative(lambda x: x**2, 3)

print(f"x=3ì—ì„œì˜ ê¸°ìš¸ê¸°: {slope:.4f}")
```

# Part 4. ì‹¤ì „ ìµœì í™”: ê²½ì‚¬í•˜ê°•ë²• (Gradient Descent)

**ìƒí™©**: $y = (x-3)^2 + 10$ ì´ë¼ëŠ” í•¨ìˆ˜(ì‚°)ê°€ ìˆìŠµë‹ˆë‹¤. ìš°ë¦¬ëŠ” $x$ë¥¼ ì¡°ì ˆí•´ì„œ $y$(ë†’ì´)ë¥¼ ìµœì†Œë¡œ ë§Œë“¤ê³  ì‹¶ìŠµë‹ˆë‹¤.

### ğŸ“ Math Example
í˜„ì¬ ìœ„ì¹˜ $x=10$ì´ë¼ê³  í•´ë´…ì‹œë‹¤.
1. **ë¯¸ë¶„**: $x=10$ì—ì„œ ê¸°ìš¸ê¸°ë¥¼ ì½ë‹ˆë‹¤. ê³µì‹ $2(x-3)$ì— ë„£ìœ¼ë©´ $2(7)=14$ì…ë‹ˆë‹¤. (ê²½ì‚¬ê°€ ì•„ì£¼ ê°€íŒŒë¥¸ ì˜¤ë¥´ë§‰)
2. **ì´ë™**: ì˜¤ë¥´ë§‰(+)ì´ë‹ˆê¹Œ ë°˜ëŒ€(-)ë¡œ ê°€ì•¼ í•©ë‹ˆë‹¤. ë³´í­(`learning_rate`)ì´ 0.1ì´ë¼ë©´,
$$ x_{new} = 10 - (0.1 \times 14) = 10 - 1.4 = 8.6 $$
3. $x$ê°€ 3ì— ê°€ê¹Œì›Œì¡ŒìŠµë‹ˆë‹¤! ì´ê±¸ ê³„ì† ë°˜ë³µí•©ë‹ˆë‹¤.

```python
def loss_func(x):
    return (x - 3)**2 + 10

current_x = 10.0
learning_rate = 0.1

for i in range(20):
    grad = numerical_derivative(loss_func, current_x)
    current_x = current_x - (learning_rate * grad)
    print(f"Step {i+1}: x = {current_x:.4f}, slope = {grad:.4f}")
```

# Part 5. ğŸ›  Library Mapping: ì‹¤ì „ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì—°ê²°

ìˆ˜í•™ ê³µì‹ê³¼ ì‹¤ì œ ë¼ì´ë¸ŒëŸ¬ë¦¬ í•¨ìˆ˜ë¥¼ 1:1ë¡œ ë§¤í•‘í•´ ë“œë¦½ë‹ˆë‹¤.

### 1. Sigmoid & Softmax (Deep Learning Activations)
- **Sigmoid**: $1 / (1 + e^{-x})$ -> `scipy.special.expit`
- **Softmax**: $e^x / \sum e^x$ -> `scipy.special.softmax`

### 2. Logarithm (ë¡œê·¸)
- `math.log`ëŠ” ê¸°ë³¸ ë°‘ì´ $e$(ìì—°ë¡œê·¸)ì…ë‹ˆë‹¤.
- `numpy`ë¥¼ ì“¸ ë•ŒëŠ” `np.log`(ìì—°ë¡œê·¸)ì™€ `np.log10`(ìƒìš©ë¡œê·¸)ë¥¼ êµ¬ë¶„í•´ì•¼ í•©ë‹ˆë‹¤.

```python
from scipy.special import expit, softmax
import numpy as np

x = np.array([1.0, 2.0, 3.0])

# Sigmoid
print(f"Sigmoid (Manual): {1 / (1 + np.exp(-x))}")
print(f"Sigmoid (Scipy) : {expit(x)}")

# Softmax
exp_x = np.exp(x)
print(f"Softmax (Manual): {exp_x / np.sum(exp_x)}")
print(f"Softmax (Scipy) : {softmax(x)}")
```
